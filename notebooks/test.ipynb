```python
import numpy as np
import polars as pl
import torch

df = pl.read_parquet(
    "../data/lichess_puzzles/part_000000000.parquet"
).slice(0, 10)
```

```python
df.head()
```

```python
decoded_df = df.select(
    pl.col("encoded_board")
    .map_elements(
        lambda x: torch.from_numpy(
            np.frombuffer(x, dtype=np.float16).astype(np.float16)
        )
    )
    .alias("decoded_board"),
    pl.col("encoded_policy_target")
    .map_elements(
        lambda x: torch.from_numpy(
            np.frombuffer(x, dtype=np.float16).astype(np.float16)
        )
    )
    .alias("decoded_policy_target"),
    pl.col("encoded_value_target")
    .map_elements(
        lambda x: torch.from_numpy(
            np.frombuffer(x, dtype=np.float16).astype(np.float16)
        )
    )
    .alias("decoded_value_target"),
)
```
