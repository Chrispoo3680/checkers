```python
import numpy as np
import polars as pl
import torch

df = pl.read_parquet(
    "../data/lichess_db_eval/lichess_db_eval.training_az73/part_000000000.parquet"
)
```

```python
decoded_df_tensor = torch.tensor(
    df.select(
        pl.col("encoded_value_target")
        .map_elements(
            lambda x: torch.from_numpy(
                np.frombuffer(x, dtype=np.float16).astype(np.float16)
            )
        )
        .alias("decoded_value_target"),
    )["decoded_value_target"].to_list()
)
```
